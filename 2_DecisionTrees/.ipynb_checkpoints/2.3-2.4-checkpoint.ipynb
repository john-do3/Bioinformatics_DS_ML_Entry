````python
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
````

````python
titanic_data = pd.read_csv("train.csv")
titanic_data.head()
````

````python
titanic_data.isnull().sum()
````

````python
X = titanic_data.drop(['PassengerId','Survived','Name', 'Ticket', 'Cabin'], axis=1)
y = titanic_data.Survived
````

Преобразуем буквенные идентификаторы в числовые, чтобы избежать ошибки scikitlearn `could not convert string to float XXX` при обучении дерева:

````python
X = pd.get_dummies(X)
````

Избавимся от `Nan` в поле `Age`, заполнив их средним значением:

````python
X = X.fillna({'Age': X.Age.median()})
````

````python
X.isnull().sum()
````

````python
clf = tree.DecisionTreeClassifier(criterion='entropy')
````

````python
clf.fit(X,y)
````

````python
import sklearn
graph = Source(sklearn.tree.export_graphviz(clf, out_file=None,
                                   feature_names=list(X),
                                   class_names=['Negative','Positive'],
                                   filled = True))
display(SVG(graph.pipe(format='svg')))
````

Неадекватный размер и структура получившегося дерева решений - следствие перобучения (overfitting).
Решение проблемы переобучения - ограничение глубины дерева или использование двух выборок.

Разобьём датасет на две последовательности - обучающую и тестовую в соотношении ~ 2/3 и 1/3:

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.33, random_state=42)
````

````python
X_train.shape
````

````python
X_test.shape
````

Оценим процент верных предсказаний для обученного дерева clf:

````python
clf.score(X, y)
````

Обучим дерево заново, используя набор, подготовленный функцией разбиения traint_test_split, затем оценим правильность предсказаний на новом наборе:

````python
clf.fit(X_train, y_train)
````

````python
clf.score(X_train, y_train)
````

Оценим правильность предсказаний на тестовом наборе данных:

````python
clf.score(X_test, y_test)
````

падение процента верных предсказаний говорит о переобучении (overfitting); попробуем исправить ситуацию ограничив глубину решений для дерева:

````python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
````

````python
clf.fit(X_train, y_train)
````

````python
clf.score(X_train, y_train)
````

````python
clf.score(X_test, y_test)
````

### эффективный подбор параметров

````python
max_depth_values = range(1, 100)
````

````python
scores_data = pd.DataFrame()
````

````python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    temp_score_data = pd.DataFrame({'max_depth': [max_depth],
                                    'train_score': [train_score],
                                    'test_score': [test_score]})
    scores_data = scores_data.append(temp_score_data)
````

Для каждого значения max_depth от 1 до 100 мы провели обучение и тесты, и добавили в датафрейм точность на тренировочной и тестовой выборке - в итоге получили 3 колонки с данными:

````python
scores_data.head()
````

Попробуем построить на графике зависимость точности предсказания от глубины обучения. Сделаем это в виде двух отдельных серий (отдельно для train и test).

Преобразуем scores_data так, чтобы для каждой строки были созданы 2 новые: одна со значением точности на тренировочных данных, другая на тестовых. Для этого задействуем встроенную функцию pandas - melt() https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.melt.html

````python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['train_score', 'test_score'],
                          var_name='set_type', value_name='score')
````

````python
scores_data_long.head()
````

- но это не всегда ведет к нахождению общих закономерностей исследуемых данных

### 5-Fold cross validation
https://scikit-learn.org/stable/modules/cross_validation.html

````python
from sklearn.model_selection import cross_val_score
````

````python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4)
````

````python
cross_val_score(clf, X_train, y_train , cv=5)
````

````python
cross_val_score(clf, X_train, y_train , cv=5).mean()
````

````python
scores_data = pd.DataFrame()
````

````python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    
    mean_cross_val_score = cross_val_score(clf, X_train, y_train , cv=5).mean()
    
    temp_score_data = pd.DataFrame({'max_depth': [max_depth],
                                    'train_score': [train_score],
                                    'test_score': [test_score],
                                    'cross_val_score': [mean_cross_val_score]})
    scores_data = scores_data.append(temp_score_data)
````

````python
scores_data.head()
````

````python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['train_score', 'test_score', 'cross_val_score'],
                          var_name='set_type', value_name='score')
````

````python
scores_data_long.head()
````

````python
sns.lineplot(x="max_depth", y="score", hue="set_type", data=scores_data_long)
````

максимальная точность достигается при большей глубине дерева, чем 3-5

Варианты:

````python
df = pd.read_csv("train_iris.csv")
X_train = df.drop(['Unnamed: 0'], axis=1)
y_train = df.species

df = pd.read_csv("test_iris.csv")
X_test = df.drop(['Unnamed: 0'], axis=1)
y_test = df.species
````

````python
import numpy as np
from sklearn.model_selection import cross_val_score

scores_data = pd.DataFrame()
rs = np.random.seed(0)

for md in range(1, 100):
    
    clf = tree.DecisionTreeClassifier(max_depth=md, random_state=rs)
    clf.fit(X_train, y_train)
    
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)

    temp_score_data = pd.DataFrame({'max_depth': [md],
                                    'train_score': [train_score],
                                    'test_score': [test_score]})
    scores_data = scores_data.append(temp_score_data)

plt.figure(figsize=(16, 6))
scores_data.plot('max_depth', ['train_score', 'test_score'])
````
